```python

    import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
from collections import defaultdict, Counter
with open("C:/Users/JRC/Desktop/survey_results_public.csv") as f:
    reader = csv.DictReader(f)
    dev_type_info = {}
    total = 0
    for data in reader:
        dev_types = data['DevType'].split(';')
        for dev_type in dev_types:
            dev_type_info.setdefault(dev_type, {
                'total': 0,
                'language_counter': Counter()
            })
            languages = data['LanguageHaveWorkedWith'].split(';')
            dev_type_info[dev_type]['language_counter'].update(languages)
            dev_type_info[dev_type]['total'] += 1


for dev_type, info in dev_type_info.items():
    print(dev_type)
    list1 = []
    list2 = []
    for language, value in info['language_counter'].most_common(10):
        language_pct = (value / info['total']) * 100
        language_pct = round(language_pct, 2)
        list2.append(language_pct) 
        print(f'----------{language}:{language_pct}%')
        list1.append(language)
    plt.title(""+dev_type+"(/)10")
    plt.xlabel("%")
    plt.ylabel("")
    plt.barh(np.arange(0,10),(list2[::-1]))
    for i in range(10):
        plt.text(x=list2[i],y=9-i,s=f'{list2[i]}%',ha="left",va="center")
    plt.yticks(np.arange(10),list1[::-1])
    plt.show()

    
```
