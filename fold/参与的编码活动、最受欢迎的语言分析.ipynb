```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
dataset = pd.read_csv("./survey_results_public.csv")
```

```python
type(dataset)
```

```python
dataset['CodingActivities'].value_counts()
```

关于CodingActivities（编码活动），大部分人都有关于爱好的活动，很小一部分在工作以外是没有编码活动的。

```python
dataset.dropna(subset=["LanguageWantToWorkWith"])
keyword_count = pd.Series(np.concatenate([str(x).lower().strip().replace('','').split(',')

for x in dataset.LanguageWantToWorkWith])).value_counts()

keyword_count = keyword_count[keyword_count != 2472]

keyword_count = keyword_count[keyword_count != 910]

result = keyword_count.nlargest(20)

result.plot(kind='bar',figsize=(12,8))

print (keyword_count)
```

从柱状图看出，Python是大家最想用的语言，可以说python是最受欢迎的语言。
